LAB 3.1 : KHÁM PHÁ DỮ LIỆU

```python
# đọc file dữ liệu
import pandas as pd

path_stocks1 = '/Users/nguyenxuanty/Documents/Lab3/DATA/stocks1.csv'

stocks1_df = pd.read_csv(path_stocks1)

# Hiển thị 5 dòng đầu tiên của dataframe stocks1_df
print('5 dòng đầu tiên của stocks1:')
print(stocks1_df.head())

# hiển thị kiểu dữ liệu trong mỗi cột trong stocks1
print('kiểu dữ liệu mỗi cột trong stocks1 là :')
print(stocks1_df.dtypes)

# xem thông tin tổng quan của stocks1
print('thông tin tổng quan của stocks1 là :')
print(stocks1_df.info())
```

LAB 3.2. XỬ LÝ DỮ LIỆU NULL

```python
#Kiểm tra xem trong stocks1 có dữ liệu Null nào không.

stocks1_df.isnull()
```

```python
#Thay thế dữ liệu Null ở cột high bằng giá trị trung bình của cột high.
mean_high = stocks1_df['high'].mean()
stocks1_df['high'].fillna(value=mean_high,inplace=True)#(inplace=True thì mặc định df sẻ thay đổi trực tiếp kh trả về bản sao mới )

#Thay thế dữ liệu Null ở cột low bằng giá trị trung bình của cột low
mean_low = stocks1_df['low'].mean()
stocks1_df['low'].fillna(value=mean_low,inplace=True)

#Hiển thị thông tin tổng quan để xác nhận không còn dữ liệu Null
print('Hiển thị thông tin tổng quan của stocks1 là :')
print(stocks1_df)
```

LAB 3.3. GỘP DỮ LIỆU VÀ PHÂN TÍCH CƠ BẢN

```python
#Đọc file stocks2.csv vào DataFrame stocks2
import pandas as pd

path_stocks2 = '/Users/nguyenxuanty/Documents/Lab3/DATA/stocks2.csv'

stocks2_df = pd.read_csv(path_stocks2)
stocks2_df
```

```python
#Gộp stocks1 và stocks2 thành DataFrame mới tên là stocks.
stocks = pd.concat([stocks1_df,stocks2_df] , ignore_index=True)#ignore_index=True là chỉ số mới sẽ được tạo lại một cách tuần tự từ 0
stocks
```

```python
#Tính giá trung bình (open, high, low, close) cho mỗi ngày
mean_values_column = stocks.groupby('date')[['open','high','low','close']].mean()
mean_values_column
```

```python
#Hiển thị 5 dòng đầu tiên của kết quả
print('5 dòng đầu tiên là :')
mean_values_column.head()
```

LAB 3.4. KẾT HỢP DỮ LIỆU TỪ NHIỀU NGUỒN

```python
#Đọc file companies.csv vào DataFrame companies.

path_companies ='/Users/nguyenxuanty/Documents/Lab3/DATA/companies.csv'

companies_df = pd.read_csv(path_companies)

# Hiển thị 5 dòng đầu tiên của companies
print('5 dòng đầu tiên của companies:')
print(companies_df.head())
```

```python
#Kết hợp stocks (đã tạo từ bài 3) và companies dựa trên cột chung là symbol.
merge_data = pd.merge(stocks,companies_df,left_on='symbol',right_on='name')
merge_data
```

```python
#Tính giá đóng cửa (close) trung bình cho mỗi công ty
mean_close = merge_data.groupby(['name'])[['close']].mean()
print('Giá đóng cửa (close) trung bình cho mỗi công ty')
print(mean_close)
```

```python
#Hiển thị kết quả cho 5 công ty đầu tiên
print('Kết quả cho 5 công ty đầu tiên')
print(mean_close.head())
```

LAB 3.5: SỬ DỤNG MULTIINDEX VÀ GROUPBY

```python
#Tạo MultiIndex cho DataFrame stocks bằng cách sử dụng cột date và symbol làm chỉ mục.
stocks.set_index(['date', 'symbol'], inplace=True)
print(stocks.head())
#set_index được sử dụng để thiết lập một hoặc nhiều cột làm chỉ mục cho DataFrame. Khi bạn thiết lập một cột làm chỉ mục, nó có nghĩa là bạn đang chọn cột đó để làm trục chỉ mục của DataFrame, thay vì sử dụng chỉ số mặc định từ 0 đến (số hàng - 1).
```

```python
#Sử dụng GroupBy để tính giá trung bình (open, high, low, close) và volume trung bình cho mỗi ngày, cho mỗi mã chứng khoán.
mean_values_stocks = stocks.groupby(level=[0, 1]).agg({'open': 'mean', 'high': 'mean', 'low': 'mean', 'close': 'mean', 'volume': 'mean'})
mean_values_stocks
#(level=[0, 1]) : chỉ định rằng chúng ta muốn nhóm dữ liệu theo hai cấp độ của chỉ mục, đó là cấp độ 0 (date) và cấp độ 1 (symbol).
#agg được sử dụng để áp dụng các hàm tổng hợp cho từng nhóm
```

```python
#Sắp xếp dữ liệu theo ngày và mã chứng khoán.
sort_values = mean_values_stocks.sort_values(by=['date','symbol'])
#Hiển thị kết quả cho 5 ngày đầu tiên
print('Hiển thị kết quả cho 5 ngày đầu tiên')
print(sort_values.head())
```

LAB 3.6: TẠO VÀ PHÂN TÍCH PIVOT TABLE

```python
#Tạo Pivot Table từ DataFrame stocks với date làm chỉ mục, symbol làm cột, và giá trị trung bình của close làm giá trị.
stocks_pivot_table = pd.pivot_table(stocks,index='date',columns='symbol',values='close',aggfunc='mean')
stocks_pivot_table
```

```python
#Thêm một cột tính tổng volume giao dịch cho mỗi mã chứng khoán (symbol)
add_columns = pd.pivot_table(stocks,index='date',columns='symbol',values=['close','volume'],aggfunc={'close':'mean','volume':'sum'})
add_columns
```

```python
#Sắp xếp Pivot Table dựa trên tổng volume giao dịch, từ cao xuống thấp
sort_pivottable = add_columns.sort_values(by='add_columns',axis=1, ascending=False)
```

```python
# Hiển thị kết quả cho 5 mã chứng khoán có tổng volume giao dịch cao nhất
top_5_stocks = sorted_pivottable.head()

# Hiển thị kết quả
print(top_5_stocks)
```
